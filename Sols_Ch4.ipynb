```python
# Exercise 4.3: Calculating derivatives

def f(x):
    return x*(x-1)

def df(x,delta):
    return (f(x+delta)-f(x))/delta

eps = 1e-14
print(df(1,eps))
```

```python
# Exercise 4.4: Calculating integrals

from math import sqrt

N = 10000
h = 2/N
integral = 0

for k in range(1,N):
    x = -1+h*k
    y = sqrt(1-x**2)
    integral += h*y

print(integral)
```
