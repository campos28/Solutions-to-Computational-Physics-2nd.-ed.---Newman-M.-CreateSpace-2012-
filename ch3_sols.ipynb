```python
# Exercise 3.6: Deterministic chaos and Feigenbaum plot 

import numpy as np
import matplotlib.pylab as plt

def feigen(r,x):
    return r*x*(1-x)

num_iterations = 1000   # Total iterations for each 'r'
last_iterations = 100   # Number of last iterations to plot (to show attractors)

# Initialize lists to store the data
x_vals = []
r_vals = []

# Iterate through each 'r' value
for r in np.arange(1,4,0.01):
    x = 0.5  # Initial condition for x

    # Iterate the feigen function
    for i in range(num_iterations):
        x = feigen(r, x)
        # Store the 'last_iterations' values for plotting
        if i >= (num_iterations - last_iterations):
            x_vals.append(x)
            r_vals.append(r)
 
plt.plot(r_vals, x_vals, ',k')
plt.title('Feigenbaum Plot')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python
# Exercise 3.7: The Mandelbrot set

import numpy as np
import matplotlib.pyplot as plt

def mandel(z,c):
    return z*z+c

max_iter = 100
w, h = 500, 500

# Sample grid
x_vals = np.linspace(-2,2,400)
y_vals = np.linspace(-2,2,400)

image = np.zeros((h, w))

for ix, x in enumerate(x_vals):
    for iy, y in enumerate(y_vals):
        c = complex(x, y)
        z = 0
        count = 0
        for k in range(max_iter):
            z = mandel(z, c)
            if abs(z) > 2:
                break
            count += 1
        image[iy, ix] = count

plt.imshow(image, cmap="hot", extent=(-2,1,-1.5,1.5))
plt.colorbar()
plt.show()               
```
